```python
import cv2
import matplotlib.pyplot as plt
path = "experiment/"
image1 = cv2.imread(path + 'hasil_perc3.jpeg', 0)
image2 = cv2.imread(path + 'hasil_perc3a.jpeg', 0)

fig=plt.figure(figsize=(15, 15))
plt.subplot(2,2,1)
plt.imshow(image1)
plt.subplot(2,2,2)
plt.imshow(image2)
```

```python
im1 = cv2.threshold(image1,127,255,cv2.THRESH_BINARY)[1]
im2 = cv2.threshold(image2,127,255,cv2.THRESH_BINARY)[1]
a = im1.sum()
b = im2.sum()
im_sum = a + b
print(a, "nilai gambar1")
print(b, "nilai gambar2")
print(im_sum)

inter = a & b
print(inter, "intersaction")
formula = ((2*inter) / im_sum )
print(formula, "DSI")
```
